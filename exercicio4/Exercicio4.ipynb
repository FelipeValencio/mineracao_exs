1- Encontre o(s) conjunto(s) com um suporte de 33% do conjunto de transações abaixo:
T1 = tree, cup, paper
T2 = book, tomato, pen
T3 = pen, book, tree
T4 = tomato, pen, cup
T5 = tree, paper, book
T6 = pen, book

```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
data = {'items': ['tree,cup,paper', 'book,tomato,pen', 'pen,book,tree', 'tomato,pen,cup', 'tree,paper,book', 'pen,book']}
 
# Create DataFrame
df = pd.DataFrame(data)

df
```

```python
data = list(df["items"].apply(lambda x:x.split(",") ))
data
```

```python
#Let's transform the list, with one-hot encoding
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
a_data = te.fit(data).transform(data)
df = pd.DataFrame(a_data,columns=te.columns_)
df
```

```python
#set a threshold value for the support value and calculate the support value.
frequent_itemsets = apriori(df, min_support=0.3, use_colnames=True)
frequent_itemsets
```

2) Dado o conjunto de transações:
Gere regras para os conjuntos relativos à linha e coluna, nessa ordem, para todas as células da
matriz abaixo. Por exemplo, 2→3, 2→5 e 2→ 7 para a primeira linha da matriz.
Informar todas as regras com confiança mínima de 73%.

```python
data = [[1,3,5,7,8], [3,4,5,7], [2,3,5,6], [2,4,5,8,9], [2,3,5,11], [1,2,3,7,9]]
data
```

```python
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
a_data = te.fit(data).transform(data)
df = pd.DataFrame(a_data,columns=te.columns_)
df
```

```python
df_sup = apriori(df, min_support=0.3, use_colnames=True)
df_sup
```

```python
#Let's view our interpretation values using the Associan rule function.
df_ar = association_rules(df_sup, metric = "confidence", min_threshold = 0.6)
df_ar
```

```python
#Filtrando regras relativos à linha e coluna
df_ar = df_ar[
    ((df_ar['antecedents'] == {2}) & (df_ar['consequents'] == {3})) |
    ((df_ar['antecedents'] == {2}) & (df_ar['consequents'] == {5})) |
    ((df_ar['antecedents'] == {2}) & (df_ar['consequents'] == {7})) |
    ((df_ar['antecedents'] == {3}) & (df_ar['consequents'] == {2})) |
    ((df_ar['antecedents'] == {3}) & (df_ar['consequents'] == {5})) |
    ((df_ar['antecedents'] == {3}) & (df_ar['consequents'] == {7})) |
    ((df_ar['antecedents'] == {5}) & (df_ar['consequents'] == {2})) |
    ((df_ar['antecedents'] == {5}) & (df_ar['consequents'] == {3})) |
    ((df_ar['antecedents'] == {5}) & (df_ar['consequents'] == {7})) |
    ((df_ar['antecedents'] == {7}) & (df_ar['consequents'] == {2})) |
    ((df_ar['antecedents'] == {7}) & (df_ar['consequents'] == {3})) |
    ((df_ar['antecedents'] == {7}) & (df_ar['consequents'] == {5}))
    ]

df_ar.loc[df_ar['confidence'] >= 0.73]
```

```python
# Utilize os dados sobre compras de supermercado “supermercado.csv”, disponível no moodle.

import csv

data = []

with open('supermercado.csv', newline='') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        data.append(row) 

data
```

```python
# Garanta que um conjunto de itens não contenha dois itens iguais.

dataf = []
for d in data:
    dataf.append(list(dict.fromkeys(d)))
    
dataf
```

```python
# Execute o Apriori com as configurações: suporte mínimo = 0.001 e confiança = 0.25
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
a_data = te.fit(dataf).transform(dataf)
df = pd.DataFrame(a_data,columns=te.columns_)

df_sup = apriori(df, min_support=0.001, use_colnames=True, verbose = 1)
df_sup
```

```python
# Execute o Apriori com as configurações: suporte mínimo = 0.001 e confiança = 0.25
df_ar = association_rules(df_sup, metric = "confidence", min_threshold=0.25)
df_ar
```

```python
# Gere regras que contenham no consequente somente itens do tipo "bottled beer"
df_ar[(df_ar['consequents'] == {'bottled beer'}) ]
```
