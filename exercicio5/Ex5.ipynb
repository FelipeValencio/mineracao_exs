```python
from sklearn.cluster import DBSCAN

df = pd.read_csv('data/X_trainToronto.csv')
df
```

```python
df_pop = df[["latitude", "longitude", "review_count"]]
df_pop
```

```python
# Ordenar o DataFrame pelo review_count
df_sorted = df_pop.sort_values(by='review_count')

# Calcular a mediana do review_count
mediana_review_count = df_sorted['review_count'].median()

# Dividir o conjunto de dados em dois grupos: mais populares e menos populares
popular_df = df_sorted[df_sorted['review_count'] >= mediana_review_count]
less_popular_df = df_sorted[df_sorted['review_count'] < mediana_review_count]
```

```python
#Para visualizar em mapas
import folium
from folium.plugins import HeatMap
```

```python
#Configuração e execução do DBSCAN
dbscanConfig = DBSCAN(eps=0.1, min_samples=25)

dbscanResults = dbscanConfig.fit(popular_df)

city_center = (43.651070, -79.347015)
map_city = folium.Map(location=city_center, zoom_start=12) 

popular_df_filtered = popular_df[dbscanResults.labels_ !=-1]

print(len(popular_df))
print(len(popular_df_filtered))

pontos = [[row[0],row[1]]  for index, row in popular_df_filtered.iterrows()]

HeatMap(pontos).add_to(map_city)
map_city
```

```python
#Configuração e execução do DBSCAN
dbscanConfig = DBSCAN(eps=0.1, min_samples=25)
dbscanResults = dbscanConfig.fit(less_popular_df)
city_center = (43.651070, -79.347015)
map_city = folium.Map(location=city_center, zoom_start=12) 

less_popular_df_filtered = less_popular_df[dbscanResults.labels_ !=-1]

print(len(less_popular_df))
print(len(less_popular_df_filtered))

pontos = [[row[0],row[1]] for index, row in less_popular_df_filtered.iterrows()]

HeatMap(pontos).add_to(map_city)
map_city
```

É possível perceber que restaurantes mais populares (com maior quantidade de reviews) tendem a estar mais no centro da cidade, e restaurantes menos populares em regiões mais periféricas. Os valores de eps e min_samples foram obtidos por teste para se obter a melhor leitura
